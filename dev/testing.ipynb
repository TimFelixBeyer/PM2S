```python
def plot_piano_roll(midi, start_time, end_time, time="beats"):
    scale_qn = 60
    start_time_qn = midi.time_to_tick(start_time) / midi.resolution * scale_qn
    end_time_qn = midi.time_to_tick(end_time) / midi.resolution * scale_qn
    pr = np.zeros((128, int((end_time_qn - start_time_qn))))

    for instrument in midi.instruments:
        for note in instrument.notes:
            if note.start >= end_time or note.end <= start_time:
                continue
            start = int((midi.time_to_tick(note.start) / midi.resolution * scale_qn - start_time_qn))
            end = int((midi.time_to_tick(note.end) / midi.resolution * scale_qn - start_time_qn))
            pr[note.pitch, start:end] = 1
    
    plt.figure(figsize=(20, 10))
    plt.subplot(2, 1, 1)
    plt.imshow(pr, aspect='auto', origin='lower', cmap='gray')
    plt.xlabel("time (quarter_notes)")

    # plot vertical lines for quarter note lines
    beats = midi.get_beats()
    for beat in beats:
        if start_time <= beat <= end_time:
            plt.axvline(x=midi.time_to_tick(beat)/midi.resolution*scale_qn, ymin=0, ymax=0.5, color='g')

    # timing based on seconds
    scale = 100
    pr = np.zeros((128, int((end_time - start_time) * scale)))

    for instrument in midi.instruments:
        for note in instrument.notes:
            if note.start >= end_time or note.end <= start_time:
                continue
            start = int((note.start - start_time) * scale)
            end = int((note.end - start_time) * scale)

            pr[note.pitch, start:end] = 1
            
    plt.subplot(2, 1, 2)
    # plot vertical lines for quarter note lines
    for beat in beats:
        if start_time <= beat <= end_time:
            plt.axvline(x=scale * beat, ymin=0.5, ymax=1, color='g')

    plt.imshow(pr, aspect='auto', origin='lower', cmap='gray')
    plt.xlabel("time (seconds)")

plot_piano_roll(mid, 0.0, 20.0)
```
